## Volem omplir les dades buides de la variable 'Age'

podem simplement substituir els valors buits per la mitjana de totes les edads
per ho podem fer millor si ens fixem en la variable Pclass

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Pclass', y='Age', data=train)
```

```python
int(train[train['Pclass']==1]['Age'].mean())
```

```python
#funci per a canviar l'edad on el valor s NULL
def imput_age(cols):
    Age = cols[0]
    Pclass = cols[1]
    
    if pd.isnull(Age):
        
        if Pclass == 1:
            return int(train[train['Pclass']==1]['Age'].mean())
        elif Pclass == 2:
            return int(train[train['Pclass']==2]['Age'].mean())
        else:
            return int(train[train['Pclass']==3]['Age'].mean())
    else:
        return Age
        
train['Age'] = train[['Age','Pclass']].apply(imput_age, axis=1)
```

```python
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
train.drop('Cabin', axis=1, inplace=True)
```

```python
sns.heatmap(train.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
train.head()
```

```python
train.dropna(inplace=True)
```

```python
train['Name'][0].split(',')[1].split('.')[0]
```

```python
train['Name'] = train['Name'].apply(lambda x: x.split(',')[1].split('.')[0])
```

```python
train['Name'].unique()
```

### Necessitem crear les dummy variables

```python
sex = pd.get_dummies(train['Sex'], drop_first=True)
#drop_first aqu es important per que si no tindriem 2 columnes completament relacionades:
# female i male, que afectarien els resultats de l'algoritme, per aix nomes en deixem una.
```

```python
sex.head(2)
```

```python
embark = pd.get_dummies(train['Embarked'], drop_first=True)
```

```python
embark.head()
```

```python
pclass = pd.get_dummies(train['Pclass'], drop_first=True)
name = pd.get_dummies(train['Name'], drop_first=True)
```

```python
train = pd.concat([train, sex, embark, pclass, name], axis=1)
```

```python
train.head(2)
```

```python
train.drop(['Sex', 'Name', 'Ticket', 'Embarked', 'Pclass'], axis=1, inplace=True)
```

```python
train.drop(['PassengerId'], axis=1, inplace=True)
```

```python
train.head()
```

```python
X = train.drop('Survived', axis=1)
y = train['Survived']
```

```python
from sklearn.cross_validation import train_test_split 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
X_train.head(2)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logmodel = LogisticRegression()
```

```python
logmodel.fit(X_train, y_train)
```

```python
predictions = logmodel.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test, predictions))
```

```python
from sklearn.metrics import confusion_matrix
```

```python
confusion_matrix(y_test, predictions)
```
